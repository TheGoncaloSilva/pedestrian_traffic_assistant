```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Plot Processing time

```python
# Filter out the models with a defined training time
filtered_results = {key: value for key, value in trainingResults.items() if "training_time" in value}

# Prepare data for plotting
epochs = []
training_times = []
labels = []

for result in filtered_results.values():
    epochs.append(result["epochs"])
    training_times.append(result["training_time"])
    label = f"{result['model']} ({result['optimizer']}, {result['lr0']}, {result['weight_decay']})"
    labels.append(label)

# Plot the data as a horizontal bar plot with number of epochs
plt.figure(figsize=(12, 6))
bar_width = 0.35
indices = range(len(epochs))

# Create horizontal bars
plt.barh(indices, training_times, height=0.35)

# Labeling
plt.ylabel('Models (Epochs)')
plt.xlabel('Training Time (hours)')
plt.title('Training Time (Platform RTX 4000, batch-size 16)')

# Combine model name and number of epochs for y-axis labels
labels_with_epochs = [f"{label} ({epochs[i]} epochs)" for i, label in enumerate(labels)]

plt.yticks(indices, labels_with_epochs)
plt.tight_layout()

# Display the plot
plt.show()
```

# Plot Inference times

```python
# Create a new map with the objects that have inference time defined
filtered_results = {key: value for key, value in trainingResults.items() if "inference_time" in value}

# Create a map with the average inference time and confidence interval for each model
inference_times = {}
for key, value in filtered_results.items():
    if value["model"] not in inference_times:
        inference_times[value["model"]] = {
            "inference_time": [value["inference_time"]],
            "conf_interval": [value["conf_interval"]]
        }
    else:
        inference_times[value["model"]]["inference_time"].append(value["inference_time"])
        inference_times[value["model"]]["conf_interval"].append(value["conf_interval"])

# Print the average inference time and confidence interval for each model
for model, values in inference_times.items():
    inference_time = np.mean(values["inference_time"])
    conf_interval = np.mean(values["conf_interval"])
    print(f"Model: {model}, Inference Time: {inference_time:.2f} ms, Confidence Interval: {conf_interval:.2f} ms")

# Bar plot for average inference time and confidence interval
# Prepare data for plotting
models = list(inference_times.keys())
avg_inference_times = [np.mean(values["inference_time"]) for values in inference_times.values()]
conf_intervals = [np.mean(values["conf_interval"]) for values in inference_times.values()]

# Plot the data
plt.figure(figsize=(12, 6))
indices = range(len(models))

# Create bars with error bars
plt.bar(indices, avg_inference_times, yerr=conf_intervals, capsize=5, width=bar_width, color='skyblue', edgecolor='black')

# Labeling
plt.xlabel('Models')
plt.ylabel('Inference Time (ms)')
plt.title('Average Inference Time with Confidence Intervals')

# Set x-ticks
plt.xticks(indices, models, rotation=45, ha='right')
plt.tight_layout()

# Display the plot
plt.show()
```
